```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import math
from PIL import Image
```

```python
img=np.array(Image.open("bicycle.bmp"))
img_gray=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
plt.imshow(img_gray)
```

```python
def generate_base_image(img,sigma,assum_blur=0.5):
    
    sigma_diff=math.sqrt(sigma**2-assum_blur**2)
    img=cv2.pyrUp(img)
    img=cv2.GaussianBlur(img,ksize=(3,3),sigmaX=sigma_diff,sigmaY=sigma_diff)
    
    return img
```

```python
base_img=generate_base_image(img_gray,1.6)
plt.imshow(base_img)
```

```python
def generate_gaussian_kernel(num_intervals,sigma):
    k=2**(1/num_intervals)
    gaus_kernel=[]
    total_img=num_intervals+3
    gaus_kernel.append(sigma)
    for i in range(total_img-1):
        prev_blur=(k**i)*sigma
        needed_blur=(k**(i+1))*sigma
        difference=round(math.sqrt(needed_blur**2-prev_blur**2),4)
        gaus_kernel.append(difference)
    
    return gaus_kernel
        
```

```python
kernel_gauss=generate_gaussian_kernel(3,1.6)
kernel_gauss
```

```python
def scale_space(base_img,gaussian_kernel,num_octaves=4):
    
    scale_space_list=[]
    
    for i in range(num_octaves):
        octave_image=[]
        octave_image.append(base_img)
        for k in gaussian_kernel[1:]:
            base_img=cv2.GaussianBlur(base_img,ksize=(3,3),sigmaX=k,sigmaY=k)
            octave_image.append(base_img)
        scale_space_list.append(octave_image)
        base_img=octave_image[-3]
        base_img=cv2.pyrDown(base_img)
    
    return scale_space_list
```

```python
scale_Space=scale_space(base_img,kernel_gauss)
```

```python
plt.imshow(scale_Space[3][-1])
```

```python
def compute_DOG(comp_space):
    
    DOG_space=[]
    
    for i in comp_space:
        DOG_octave=[]
        for j in range(len(i)-1):
            image=i[j+1]-i[j]
            DOG_octave.append(image)
        
        DOG_space.append(DOG_octave)
    
    return DOG_space
```

```python
DOG=compute_DOG(scale_Space)
```
